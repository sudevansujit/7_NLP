```python
from nltk import FreqDist, word_tokenize
from nltk.util import ngrams
```

```python
mytext = open('legendofsleepyhollow.txt', 'r').read()
len(mytext)
```

```python
mytext
```

```python
mysplit = mytext[:300]
mysplit
```

```python
mysplit.replace('\n', ' ' )
```

```python
mytoken = word_tokenize(mysplit)
```

```python
print(mytoken)
```

#### Freq Dist of 1 grams

```python
one_gram = list(ngrams(mytoken, 1, pad_left= True,
                                        pad_right=True,
                                        left_pad_symbol = "_",
                                        right_pad_symbol = "_"))


fDist_one = FreqDist(one_gram)
fDist_one
```

```python
fDist_one[('of', )]  # We have to change the key format
```

```python
unigram_dict = dict()
for k, v in fDist_one.items():
        unigram_dict[' '.join(k).lower()] = v
print(unigram_dict)
```

```python
# Arranging in descending order
from collections import Counter

c = Counter(unigram_dict)
print(c)
```

### Applying it on Book - legend of sleepy hollow

```python
token2 = word_tokenize(mytext)
```

```python
mygram = list(ngrams(token2, 1,  pad_left= True,
                                        pad_right=True,
                                        left_pad_symbol = "_",
                                        right_pad_symbol = "_"))

fDist_mygram = FreqDist(mygram)
my_dict = {}
for k, v in fDist_mygram.items():
    my_dict[' '.join(k).lower()] = v
    
final_dict = Counter(my_dict)
print(final_dict)
    

    
```

```python
import operator
english_unigram_freqs = sorted(final_dict.items(), key=operator.itemgetter(1), reverse=True)
english_unigram_freqs[:10]
```

```python
import numpy as np
import matplotlib.pyplot as plt
labels, values = zip(*english_unigram_freqs[:10])
indexes = np.arange(len(labels))
width = 0.8 # width = 1 would give bars that overlap because they are too close.

fig = plt.figure(figsize=(10,7))                                                               
ax = fig.gca() # Get current axis
rects = ax.bar(indexes, values, width)

# Add title and axis labels
fig.suptitle('Top 10 English word unigrams', fontsize=20)
plt.xlabel('Word unigram', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Display value of each bar on bar
for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., 50 + height, '%d' % int(height), ha='center', va='bottom') # Can also add color and fontweight arguments.

# Remove the default x-axis tick numbers and use tick numbers of your own choosing:
ax.set_xticks(indexes)
# Replace the tick numbers with strings:
ax.set_xticklabels(labels)

plt.show()
# plt.savefig('top10EnglishWordUnigrams.png')
```
