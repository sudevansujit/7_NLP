# Visualising Using Displacy

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc_covid = nlp(open('covid19.txt').read())
doc_covid
```

```python
for token in doc_covid:
    print('{:<20},{:<20},{:<20}'.format(token.text,token.pos_, token.dep_))
```

```python
from spacy import displacy
```

```python
displacy.render(doc_covid, style = 'dep')
```

```python
options = {'compact': True, 'bg': 'cornflowerblue', 
            'color' : '#fff', 'font': 'Sans Serif'}
```

```python
displacy.render(doc_covid, style = 'dep', options = options)
```

### Saving Figure

```python
save_fig = displacy.render(doc_covid, style = 'dep', options = options)
```

```python
with open('doc1.svg', 'w', encoding = 'utf-8') as f:
    f.write(save_fig)
```
